**heirarchical clustering:**

 In this technique, initially each data point is considered as an individual cluster. At each iteration, the similar clusters merge with other clusters until one cluster or K clusters are formed. steps are:

* Compute the similarity matrix (Using needed parameters. i.e. euclidean distance)
* Let each data point be a cluster
* Repeat: Merge the two closest clusters and update the simularity matrix and dendogram
* Until only a single cluster remains
* Cut the dendogram to find the k clusters as needed

Calculating the similarity between two clusters is important to merge or divide the clusters. There are certain approaches which are used to calculate the similarity between two clusters:






Distance Between Centroids


* MIN
  
  sets the similarity between two closest points in two clusters as the similarity of them:

  Mathematically this can be written as,

  Sim(C1,C2) = Min Sim(Pi,Pj) such that Pi ∈ C1 & Pj ∈ C2 (Similarity function could change as needed)

```python
from IPython.display import Image

Image(url = "https://miro.medium.com/v2/resize:fit:640/format:webp/1*mtDL2TynaiwpJlhLdecFYQ.jpeg")

#Reference: https://towardsdatascience.com/understanding-the-concept-of-hierarchical-clustering-technique-c6e8243758ec
```

* MAX
  
  sets the similarity between two furthest points in two clusters as the similarity of them:

  Mathematically this can be written as,

  Sim(C1,C2) = Max Sim(Pi,Pj) such that Pi ∈ C1 & Pj ∈ C2 (Similarity function could change as needed)

```python
from IPython.display import Image

Image(url = "https://miro.medium.com/v2/resize:fit:640/format:webp/1*nRYZyjoT1ZRzlWp3oP0_QQ.jpeg")

#Reference: https://towardsdatascience.com/understanding-the-concept-of-hierarchical-clustering-technique-c6e8243758ec
```

* Group Average
  
  Take all the pairs of points and compute their similarities and calculate the average of the similarities.

  Mathematically this can be written as,

  sim(C1,C2) = ∑ sim(Pi, Pj)/|C1|*|C2|. where, Pi ∈ C1 & Pj ∈ C2 (Similarity function could change as needed)

* Note: there is also an algorithm named "ward method" which does the exact process but it uses the square of distances.

**DBSCAN**

DBSCAN is a density based clustering algorithm.

But what is Density-Based Clustering?

* In density based clustering we partition points into dense regions separated by not-so-dense regions.

The DBSCAN algorithm uses two major parameters:

* minPts: The minimum number of data points that can form a cluster.
* Dense Region: A circle of radius Eps that contains at least
MinPts points.

Characterization of points:
* A point is a core point if it has more than a
specified number of points (MinPts) within a circle with radius of Eps
* A border point has fewer than MinPts within
Eps, but is in the neighborhood of a core point.
* A noise point is any point that is not a core point
or a border point.

now we must talk about two important concepts:

Density edge
* We place an edge between two
core points q and p if they are
within distance Eps.

Density-connected
* A point p is density-connected to a
point q if there is a path of edges
from p to q

now let's explain the algorithm:

1. Label points as core, border and noise
1. Eliminate noise points
2. For every core point p that has not been
assigned to a cluster:
* Create a new cluster with the point p and all the
points that are density-connected to p.
4. Assign border points to the cluster of the
closest core point.

#####**K-means:**#####

in this section we compute the k-means clustering for up to 40 clusters and plot SSE for each number of clusers. as you can see in below's plot, best number of clusters by elbow method seems to be around 5, 6 or 7 clusters. we also want to use silhoute score for a better decision.

```python
distorsions = []
max_loop=40
for k in range(2, max_loop):
    k_means = KMeans(n_clusters=k)
    k_means.fit(X)
    distorsions.append(k_means.inertia_)
fig = plt.figure(figsize=(10, 5))
plt.plot(range(2, max_loop), distorsions)
plt.xticks([i for i in range(2, max_loop)], rotation=75)
plt.xlabel("Number of clusters")
plt.ylabel("Sum of Square Error")
plt.grid(True)
```

here we do the same but this time we compute silhoute score for each cluster. we can say that 5 clusters has a good sillhoute score comparing with 6 and 7. also 5 is a better elbow point in both plots, therefore 5 is our choice.

```python
from sklearn import metrics

silhouette_score = []
for k in range(2, max_loop):
        kmeans = KMeans(n_clusters=k,  random_state=10, n_init=10)
        kmeans.fit(X)
        silhouette_score.append(metrics.silhouette_score(X, kmeans.labels_, random_state=10))
fig = plt.figure(figsize=(10, 5))
plt.plot(range(2, max_loop), silhouette_score)
plt.xticks([i for i in range(2, max_loop)], rotation=75)
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette score")
plt.grid(True)
```

#####**Hierarchical:**#####

for this method we will use AgglomerativeClustering from sklearn.cluster. we will use defferent types of linkages and compare their sillhoute score to choose the best linkage. also the number of clusters here will be 5 as we want to compare with our k-means model.

```python
from sklearn.cluster import AgglomerativeClustering

for the_linkage in ("ward", "average", "complete", "single"):
    clustering = AgglomerativeClustering(linkage=the_linkage, n_clusters=5)
    clustering.fit(X)
    labels = clustering.labels_
    sc = metrics.silhouette_score(X, labels)
    print(f"Silhouette Coefficient for {the_linkage} is:%0.2f" % sc)

```

as you can see the best method is average with sillhoute score of 0.19, which is almost around k-means method (just a little bit better). we also plot it's dendogram:

```python
from scipy.cluster.hierarchy import linkage, dendrogram

L = linkage(X, method="average", metric="euclidean")

d = dendrogram(L)
```

#####**DBSCAN:**#####

In order to determine the best value of eps for our dataset, we should use the K-Nearest Neighbours approach as explained in these two papers: Sander et al. 1998 and Schubert et al. 2017 (both papers from the original DBSCAN authors, link in references).

The approach is: If you have N-dimensional data to begin, then choose n_neighbors in sklearn.neighbors.NearestNeighbors to be equal to 2xN - 1, and find out distances of the K-nearest neighbors (K being 2xN - 1) for each point in your dataset. Sort these distances out and plot them to find the "elbow" which separates noisy points (with high K-nearest neighbor distance) from points (with relatively low K-nearest neighbor distance) which will most likely fall into a cluster. The distance at which this "elbow" occurs is your point of optimal eps.

Also it was mentioned that the best number of min points is twice the number of features. therefore we first plot the explained graph to find the optimum value of eps:

```python
from sklearn.neighbors import NearestNeighbors

def get_kdist_plot(X=None, k=None):

    nbrs = NearestNeighbors(n_neighbors=k).fit(X)

    # For each point, compute distances to its k-nearest neighbors
    distances, indices = nbrs.kneighbors(X)

    distances = np.sort(distances, axis=0)
    distances = distances[:, k-1]

    # Plot the sorted K-nearest neighbor distance for each point in the dataset
    plt.figure(figsize=(8,8))
    plt.plot(distances)
    plt.xlabel('Points/Objects in the dataset', fontsize=12)
    plt.ylabel('Sorted {}-nearest neighbor distance'.format(k), fontsize=12)
    plt.grid(True, linestyle="--", color='black', alpha=0.4)
    plt.show()
    plt.close()


k = 2 * X.shape[-1] - 1
get_kdist_plot(X=X, k=k)
```

Therefore by looking at graph the eps seems to be somewhere around 1.75 and 2.5. we will test all:

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=2.5, min_samples=24).fit(X)
labels = db.labels_
classes, repeats = np.unique(np.array(labels), return_counts=True)
print(f"classes are: {classes} and repeats are: {repeats}")

n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
print(n_clusters_)

sc = metrics.silhouette_score(X, labels)
print("Silhouette Coefficient:%0.2f" % sc)
```

eps = 2.5 has good sillhoute score, but it made only one cluster and that couldn't be acceptable.
